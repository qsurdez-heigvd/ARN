# üê≠ &nbsp; <font color="#6DA4AA">Electroencephalogram of Sleeping Mice - Data Exploration</font>

In this practical work, we will explore a dataset about mice's sleep obtained from CHUV. This dataset includes the Electroencephalogram (EEG) of a mouse, as well as its state.

In the long run the objective would be to be able to recognize mice sleep stages from their EEG.

![Sleeping mouse](https://www.cottagesmallholder.com/wp-content/images/mouse-asleep-with-tiny-teddy-bear.jpg)

EEG is a scientific tool that helps us study brain activity. It records the electrical signals in the brain over time, creating a time serie with detailed patterns. These patterns include things like alpha oscillations suggesting a relaxed but awake state, beta oscillations that might indicate active thinking and focus, or theta oscillations typically observed during deep sleep and non-REM sleep stages.

As a reminder, the frequency in Hertz is the number of cycles per second:

![Frequency](https://cmtext.indiana.edu/acoustics/img/4cps.png)

The problem is that the information about these patterns is spread out over many data points, making it hard to detect directly the main frequency of such oscillations.

So, we use something called spectral analysis, which transforms these time series into a different view called the frequency domain. This helps us see the main patterns more clearly by concentrating information about one frequency of oscillation. The process of converting data from the time-domain to the frequency-domain is done thanks to Fourier transforms.

See this diagram to better understand the relationship between the time domain and the frequency domain:
![Frequency Domain](https://www.dataphysics.com/wp-content/uploads/2022/01/The_Frequency_Domain_2.jpg)

Let's look a bit closer at our data. A row was generated each 4 seconds (=15 by minute) during one entire day, giving a dataset of 15x60x24=21600 datapoints. After removing outliers, we are left with 20258 datapoints. For each of this datapoint, the mouse is either:

- awake (**w**)
- in REM sleep (**r**)
- in non-REM sleep (**n**).

In addition to its current state, each datapoint (row) is characterized with a "spectrum". As mentioned earlier, it is really difficult to use the raw electrical activity of the mouse brain. This is why we will focus on the spectrum in order to predict the state of the mouse. The spectrum is described as the amplitude of the signal at each frequency between 1 and 101 [Hz] during the last 4 seconds (=since the last datapoint).

The main point to understand is that at each timestamp (= row = datapoint), the mouse is in a specific state and its brain electrical activity is represented by 101 different features based on frequencies.

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

# Calculate state durations
def get_state_durations(state_series):
    durations = defaultdict(list)
    current_state = state_series.iloc[0]
    current_duration = 4
    
    for i in range(1, len(state_series)):
        if state_series.iloc[i] == current_state:
            current_duration += 4
        else:
            durations[current_state].append(current_duration)
            current_state = state_series.iloc[i]
            current_duration = 4
            
    # Add the last duration
    durations[current_state].append(current_duration)
    return durations

# Calculate durations
durations = get_state_durations(mice_eeg.state)

# Normalize the data using min-max normalization
def normalize_data(data_list):
    min_val = min(data_list)
    max_val = max(data_list)
    return [(x - min_val) / (max_val - min_val) if max_val != min_val else 0 for x in data_list]

# Normalize each state's durations
normalized_data = [normalize_data(durations[state]) for state in ['w', 'n', 'r']]

# Create two subplots side by side
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 6))

# Original data plot
ax1.boxplot([durations['w'], durations['n'], durations['r']], 
         tick_labels=['Wake', 'Non-REM', 'REM'])
ax1.set_title('Original Duration Distributions')
ax1.set_ylabel('Duration (seconds)')
ax1.set_xlabel('State')
ax1.grid(True, alpha=0.3)

# Log scale plot
ax2.boxplot([durations['w'], durations['n'], durations['r']], 
         tick_labels=['Wake', 'Non-REM', 'REM'])
ax2.set_title('Log scaled Duration Distributions')
ax2.set_yscale('log')
ax2.set_ylabel('Duration (seconds)')
ax2.set_xlabel('State')
ax2.grid(True, alpha=0.3)

# Normalized violin plot
ax3.violinplot(normalized_data)#, labels=['Wake', 'Non-REM', 'REM'])
ax3.set_title('Normalized Duration Distributions')
ax3.set_ylabel('Normalized Duration (0-1)')
ax3.set_xlabel('State')
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()


# Print statistics for both original and normalized data
for state, norm_data in zip(['w', 'n', 'r'], normalized_data):
    print(f"\nState {state}:")
    print(f"Original - Longest duration: {max(durations[state])} seconds")
    print(f"Original - Median duration: {np.median(durations[state])} seconds")
    print(f"Normalized - Median: {np.median(norm_data):.3f}")
```

6.a: I chose 1, 3 and 8 Hz. It seems that they are the most discriminative ones for the different states.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score

data = mice_eeg
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

colors = {'w': 'blue', 'n': 'red', 'r': 'green'}
labels = {'w': 'Wake', 'n': 'non-REM', 'r': 'REM'}

for state in ['w', 'n', 'r']:
    mask = data['state'] == state
    ax.scatter(
        data.loc[mask, 'amplitude_around_1_Hertz'],
        data.loc[mask, 'amplitude_around_3_Hertz'],
        data.loc[mask, 'amplitude_around_8_Hertz'],
        c=colors[state],
        label=labels[state],
        alpha=0.6
    )

ax.set_xlabel('Amplitude around 1 Hz', labelpad=8)
ax.set_ylabel('Amplitude around 3 Hz', labelpad=8)
ax.set_zlabel('Amplitude around 8 Hz', labelpad=8)
ax.set_title('3D Scatter Plot of Sleep States')
ax.legend()
plt.show()
```

6.b: They are not very distinguishable as the states are quite overlapped with each other

```python
def classify(features):
  if features['amplitude_around_1_Hertz'] > 0.0000125:
    return 'w'
  elif features['amplitude_around_8_Hertz'] > 0.00002:
    if features['amplitude_around_3_Hertz'] < 0.00002:
      return 'r'
    else:
      return 'n'
  else:
    return 'w'
```

```python
selected_amplitudes = [
  'amplitude_around_1_Hertz',
  'amplitude_around_3_Hertz',
  'amplitude_around_8_Hertz'
]

pred_class = pred_class = [classify(f) for k, f in mice_eeg[selected_amplitudes].iterrows()]
```

```python
print(f"Accuracy: {np.round(np.count_nonzero(pred_class == mice_eeg.state) / len(pred_class) * 100, 2)}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Calculate confusion matrix
labels = ['w','r','n']
cm = confusion_matrix(mice_eeg.state, pred_class, labels=['w','r','n'])

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xticks([0.5, 1.5, 2.5], labels)
plt.yticks([0.5, 1.5, 2.5], labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(mice_eeg.state, pred_class, average='macro')
print(f"F1 Score: {np.round(f1, 2)}")
```
